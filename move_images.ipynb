# Image Cleanup and Relocation

This notebook performs the following steps:

1. Lists and deletes image files (excluding JavaScript files) from the root directory.
2. Moves the deleted images to the `images` directory, ensuring the directory exists.

> **Note:** The files will be backed up temporarily before moving to avoid unintentional data loss.

```python
# Section 1: Import Required Libraries
import os
import shutil
from pathlib import Path

# Define base directory
base_dir = Path('/workspaces/pizzatower')
backup_dir = base_dir / 'backup_images'
images_dir = base_dir / 'images'

# Ensure backup directory exists
backup_dir.mkdir(exist_ok=True)
```

```python
# Section 2: List and Delete Image Files (Excluding JS Files) in Main Directory

# Define image extensions
image_exts = {'.png', '.jpg', '.jpeg', '.gif', '.bmp', '.svg'}
deleted_files = []

for item in base_dir.iterdir():
    if item.is_file():
        if item.suffix.lower() in image_exts:
            # back up and delete
            dest = backup_dir / item.name
            shutil.copy2(item, dest)
            item.unlink()
            deleted_files.append(dest)

print(f"Backed up and removed {len(deleted_files)} image files.")
```

```python
# Section 3: Move Deleted Images Back to Images Directory

# Ensure images directory exists
images_dir.mkdir(exist_ok=True)

for backup_path in deleted_files:
    dest_path = images_dir / backup_path.name
    shutil.move(str(backup_path), dest_path)

print(f"Moved {len(deleted_files)} images to {images_dir}.")
```
